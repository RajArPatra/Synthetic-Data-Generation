Save_Model

```
def save_checkpoint(model, save_folder, name, epoch, opt, best_model=False):

    if best_model:
        save_path = "{}/{}/best_model.th".format(
            save_folder, name)
    else:
        save_path = "{}/{}/epoch_{}.th".format(save_folder,
                                               name, epoch)

    if not os.path.exists(save_folder + "/" + name):
        os.makedirs(save_folder + "/" + name)

    print("SAVE MODEL to", save_path)


    param_dict = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': opt.state_dict()}
        

    torch.save(param_dict, save_path)
```

Load_model

```
def load_model(load_path,model,opt):

    checkpoint = torch.load(load_path)

    epoch = checkpoint['epoch']
   

    model =model # code for model init
    model.load_state_dict(checkpoint['model_state_dict'])
    

    opt =opt #code for opt init
    opt.load_state_dict(checkpoint['optimizer_state_dict'])
    #opt.param_groups[0]['lr'] = checkpoint['optimizer_params']['lr']
    


    return model, opt, epoch
```

```
class Fc(nn.Module):
 def __init__(self,in_features,out):
    super(Fc,self).__init__()
    self.l=nn.Sequential(nn.LayerNorm(in_features,),nn.Linear(in_features,out),
                               nn.ReLU())
 def forward(self,x):
   return self.l(x) 
```

```
class encoder(nn.Module):
 def __init__(self,in_features):
    super(encoder,self).__init__()
    self.layers=nn.Sequential(nn.Linear(in_features,320),
                               nn.ReLU(),
                              nn.Linear(320,160),
                               nn.ReLU(),
                                nn.Linear(160,80),
                               nn.ReLU(),
                                nn.Linear(80,40), nn.ReLU(),
                                nn.Linear(40,20),nn.ReLU(),
                                nn.Linear(20,10), nn.ReLU())
    #self.bottle_neck=nn.Linear(10,1)
 def forward(self,x):
   return self.layers(x)
```

```
class decoder(nn.Module):
 def __init__(self,out_features):
    super(decoder,self).__init__()
    self.layers=nn.Sequential( nn.Linear(10,20),
                              nn.ReLU(),
                              nn.Linear(20,40),nn.ReLU(),
                              nn.Linear(40,80),
                               nn.ReLU(),
                              nn.Linear(80,160),
                               nn.ReLU(),
                              nn.Linear(160,320),
                               nn.ReLU(),
                           nn.Linear(320,out_features), nn.ReLU())
    
 def forward(self,x):
   return self.layers(x)
```

```
class Autoencoder(nn.Module):
  def __init__(self,features):
    super(Autoencoder,self).__init__()
    self.layers=nn.Sequential(encoder(features),decoder(features))
  def forward(self,x):
    return self.layers(x)  
```

```
class Lstm_classifier(nn.Module):
  def __init__(self,in_length,out):
    super(Lstm_classifier,self).__init__()
    self.enc=encoder(in_length)
    self.Lstm=nn.LSTM(1,32,2,batch_first=True,bidirectional=True)
    self.fc1=nn.Linear(640,64)
    self.fc2=nn.Linear(64,4)
    self.soft= nn.LogSoftmax(dim=1)
  def forward(self,x):
    x=self.enc(x)
    x=torch.transpose(x,1,2)
    out,(hn,cn)=self.Lstm(x)
    out=torch.transpose(out,1,0)
    out=self.soft(self.fc2(self.fc1(out.view(-1,10*64))))
    return out
```

```
class classifier(nn.Module):
  def __init__(self,in_length,out):
    super(classifier,self).__init__()
    self.fc0=Fc(in_length,256)
    self.fc1=Fc(256,64)
    self.fc2=Fc(64,4)
    self.soft= nn.LogSoftmax(dim=1)
    self.i=in_length
  def forward(self,x):
    x=self.fc0(x.view(-1,self.i))  
    out=self.soft(self.fc2(self.fc1(x)))
    return out
```
